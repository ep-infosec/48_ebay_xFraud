```python
def PolynomialRegression(degree=2, **kwargs):
    return make_pipeline(PolynomialFeatures(degree), LinearRegression(**kwargs))
```

```python
def avg_hitrate_train(y_true, y_pred):
    max_mean = np.mean(y_pred).max()
    return max_mean
```

```python
param_grid = {'polynomialfeatures__degree': np.arange(10), 
              'linearregression__fit_intercept': [True, False], 'linearregression__normalize': [True, False]}
poly_grid = GridSearchCV(PolynomialRegression(), param_grid, 
                         #scoring=make_scorer(avg_hitrate_train, greater_is_better=True), 
                         scoring='neg_mean_squared_error',
                         n_jobs = -1,
                         verbose=3) 
```
